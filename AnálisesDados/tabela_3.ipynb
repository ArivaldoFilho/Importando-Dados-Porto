```python
import pandas as pd
```

```python
relatorio = pd.read_html(r'https://www.gov.br/infraestrutura/pt-br/assuntos/transporte_aquaviario/sistema-portuario', header=1)[2]
display(relatorio)
```

```python
colunas_porto = relatorio['Porto'].tolist() + relatorio['Porto.1'].tolist()
colunas_estado = relatorio['Estado'].tolist() + relatorio['Estado.1'].tolist()
colunas_autoridadeportuaria = relatorio['Autoridade Portuária'].tolist() + relatorio['Autoridade Portuária.1'].tolist()

relatorio = pd.DataFrame({'Porto': colunas_porto, 'Estado': colunas_estado, 'Autoridade Portuária': colunas_autoridadeportuaria})

display(relatorio)
```

```python
relatorio = relatorio.drop(index=17)
display(relatorio)
```
