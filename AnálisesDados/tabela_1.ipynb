```python
import pandas as pd
```

```python
relatorio = pd.read_html(r'https://www.gov.br/infraestrutura/pt-br/assuntos/transporte_aquaviario/sistema-portuario', header = 1) [0] 
df_relatorio = pd.DataFrame(relatorio)
display(relatorio)
```

```python
df_relatorio = df_relatorio.drop('Unnamed: 6', axis = 1)
display(df_relatorio)
```

```python
colunas_nome = df_relatorio['Nome'].tolist() + df_relatorio['Nome.1'].tolist() + df_relatorio['Nome.2'].tolist()
colunas_estado = df_relatorio['Estado'].tolist() + df_relatorio['Estado.1'].tolist() + df_relatorio['Estado.2'].tolist()

df_relatorio = pd.DataFrame({'Nome': colunas_nome, 'Estado': colunas_estado})

display(df_relatorio)
```

```python
df_relatorio['Nome'] = df_relatorio['Nome'].str.replace(r'^\d+\s+', '', regex=True)
df_relatorio['Nome'] = df_relatorio['Nome'].str.strip('-')

# Exibir o DataFrame atualizado
display(df_relatorio)
```

```python
titulo = "Relatório Portos Fluviais"
df_relatorio = df_relatorio.style.set_caption(titulo)

display(df_relatorio)
```

```python
diretorio = r'C:\Users\ariva\Documents\projetosVsCode\pbl\\'
nome_arquivo = 'relatorioPortosFluviais.txt'
caminho_completo = diretorio + nome_arquivo

df_relatorio_string = df_relatorio.to_string()
with open(caminho_completo, 'w') as arquivo:
    arquivo.write(df_relatorio_string)
    
df_relatorio.to_
    
print(f"DataFrame exportado para {caminho_completo} com sucesso!")
```
